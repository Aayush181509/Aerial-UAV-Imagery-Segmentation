```python
import rasterio
import matplotlib.pyplot as plt

# Replace 'images/interest.tif' with the correct path to your GeoTIFF file
with rasterio.open('images/interest.tif') as image:
    image_array = image.read()

# Display each band separately
for band_number in range(image_array.shape[0]):
    plt.figure(figsize=(8, 8))
    plt.imshow(image_array[band_number, :, :], cmap='gray')  # Use 'gray' colormap for single-band images
    plt.title(f'Band {band_number + 1}')
    plt.show()

# Display a composite color image (assuming RGB)
rgb_image = image_array[:3, :, :].transpose(1, 2, 0)  # Assuming bands are in order Red, Green, Blue
plt.figure(figsize=(8, 8))
plt.imshow(rgb_image)
plt.title('RGB Composite')
plt.show()
```
