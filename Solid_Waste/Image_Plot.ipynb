```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def split_image(image, tile_size):
    height, width, _ = image.shape
    tiles = []

    for y in range(0, height, tile_size):
        for x in range(0, width, tile_size):
            tile = image[y:y+tile_size, x:x+tile_size]
            tiles.append(tile)

    return tiles
```

```python
def calculate_average_color(tile):
    return np.mean(tile, axis=(0, 1))
```

```python
def plot_rgb_graph(rgb_values):
    plt.figure(figsize=(len(rgb_values), 3))
    plt.imshow([rgb_values], aspect='auto', extent=(0, len(rgb_values), 0, 1))
    plt.xlabel('Tile Number')
    plt.ylabel('RGB Value')
    plt.title('RGB Values for Each Tile')
    plt.show()
```

```python
image_path = 'dataset/images/tile_36.jpg'
tile_size = 50  # Adjust this according to your preference

image = cv2.imread(image_path)
tiles = split_image(image, tile_size)

rgb_values = [calculate_average_color(tile) for tile in tiles]
plt.imshow(image)
plot_rgb_graph(rgb_values)
```
