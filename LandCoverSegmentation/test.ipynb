```python
import numpy as np
import matplotlib.pyplot as plt

# Simulating training and validation data with subtle fluctuations
np.random.seed(42)
epochs = 100
start_iou = 45.0
start_loss = 2.0  # Some initial value for loss

# Simulating IOU scores with subtle fluctuations
train_iou_base = start_iou + 25 * np.exp(-np.arange(epochs) / 20)
val_iou_base = start_iou + 15 * np.exp(-np.arange(epochs) / 25)

train_iou = train_iou_base + 2 * np.sin(np.arange(epochs) / 5) + np.random.normal(scale=1, size=epochs)
val_iou = val_iou_base + 1.5 * np.sin(np.arange(epochs) / 8) + np.random.normal(scale=1, size=epochs)

# Simulating loss with subtle fluctuations
train_loss_base = start_loss + 1.0 * np.exp(-np.arange(epochs) / 30) * np.sin(np.arange(epochs) / 5)
val_loss_base = start_loss + 0.8 * np.exp(-np.arange(epochs) / 25) * np.sin(np.arange(epochs) / 8)

train_loss = train_loss_base + 0.3 * np.random.normal(scale=1, size=epochs)
val_loss = val_loss_base + 0.2 * np.random.normal(scale=1, size=epochs)

# Plotting the curves
plt.figure(figsize=(12, 6))

# Plotting IOU curves
plt.subplot(1, 2, 1)
plt.plot(range(epochs), train_iou, label='Training IOU')
plt.plot(range(epochs), val_iou, label='Validation IOU')
plt.title('IOU Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('IOU (%)')
plt.legend()

# Plotting loss curves
plt.subplot(1, 2, 2)
plt.plot(range(epochs), train_loss, label='Training Loss')
plt.plot(range(epochs), val_loss, label='Validation Loss')
plt.title('Loss Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()
```
